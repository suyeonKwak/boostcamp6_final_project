```python
import pandas as pd
import json
import os
import tqdm
from collections import defaultdict
os.getcwd()
```

# build dataset

```python
for i in range(10):
    with open(f'./summary/art/CNTS-00047966809_{i}.json') as file:
        data = json.load(file)

    for k,v in data.items():
        if isinstance(v,dict):
            print(f'{k} :')
            for sub_k, sub_v in dict(v).items():
                print(f'    {sub_k} : {sub_v}')
        else:
            print(f'{k} : {v}')
    print()
```

```python
folder_path = './summary'
category_path = os.listdir(folder_path)
category_path
```

```python
info = defaultdict(list)
for c_path in category_path:
    data_path = os.listdir(folder_path + '/' + c_path)
    for path in tqdm.tqdm(data_path):
        with open(folder_path + '/' + c_path +'/'+path) as file:
            data = json.load(file)

        info['id'].append(data['passage_id'])
        info['passage'].append(data['passage'])
        info['summary'].append(data['summary'])
        info['doc_id'].append(data['metadata']['doc_id'])
        info['year'].append(data['metadata']['published_year'])
        info['label'].append(data['metadata']['kdc_label'])
```

```python
df = pd.DataFrame(info)
display(df.head())
df.info()
```

```python
df['passage_len'] = df['passage'].apply(len)
df['summary_len'] = df['summary'].apply(len)
```

```python
df.describe()
```

```python
df.to_csv('./summary.csv',index=False)
```
